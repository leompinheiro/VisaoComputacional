```python
# Importar bibliotecas
import cv2
import numpy as np
from skimage import io
from google.colab.patches import cv2_imshow
import matplotlib.pylab as plt
```

```python
#---
# Exercício 01: Ler uma imagem hospedada na Internet e exibi-la no projeto. (0,30 ponto)
#---

# Carregando imagens da internet

#img = io.imread('https://user-images.githubusercontent.com/125405197/219055718-62ef5022-53e1-4052-b6da-e28936c740a2.JPG')
img = io.imread('https://user-images.githubusercontent.com/125405197/219055872-a2276b18-9396-4539-abba-7a4c9c470c14.jpg')
imgE1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Exibindo a imagem original
cv2_imshow(imgE1)
```

```python
#---
# Exercício 02: Exibir informações sobre a imagem, tais como Altura, Largura e Quantidade de Canais. (0,30 ponto)
#---

# Exibindo informações sobre a imagem
print ("Altura: %d pixels" % (imgE1.shape[0]))
print ("Largura: %d pixels" % (imgE1.shape[1]))
print ("Quantidade de Canais: %d" % (imgE1.shape[2]))
```

```python
#---
# Exercício 03: Redimensionar a imagem para uma resolução de 200 X 200 pixels. (0,30 ponto)
#---

# redimensionando a imagem
imgE3 = cv2.resize(imgE1,(200,200), interpolation = cv2.INTER_AREA)

# Exibindo a imagem redimensionada
cv2_imshow(imgE3)

# Exibindo informações da imagem redimensionada
print ("Altura: %d pixels" % (imgE3.shape[0]))
print ("Largura: %d pixels" % (imgE3.shape[1]))
print ("Quantidade de Canais: %d" % (imgE3.shape[2]))
```

```python
#---
# Exercício 04: Aplicar um filtro Gaussiano na imagem do exercício 03 e exibir o resultado. (0,30 ponto)
#---

# Aplicando filtro Gaussiano
imgE4=cv2.GaussianBlur(src=imgE3,ksize=(7,7),sigmaX=3)

# Exibindo a imagem após a aplicação do filtro Gaussiano
cv2_imshow(imgE4)
print ("Altura: %d pixels" % (imgE4.shape[0]))
print ("Largura: %d pixels" % (imgE4.shape[1]))
print ("Quantidade de Canais: %d" % (imgE4.shape[2]))
```

```python
#---
# Exercício 05: Aplicar uma limiarização na imagem do exercício 03 e exibir o resultado. (0,30 ponto)
#---

# Convertendo para escala de cinza
imgE5=cv2.cvtColor(imgE3,cv2.COLOR_RGB2GRAY)

# Exibindo imagem em escala de cinza
cv2_imshow(imgE5)

# Construindo a máscara
mascara = imgE5 > 87
imgE5[mascara] = 255 

# Exibindo o resultado
plt.imshow(imgE5, cmap='gray')
```
